```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("C:/Users/gpro0/OneDrive/Рабочий стол/train.csv")
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
for i in df.select_dtypes(['object']).columns:
    df[i] = df[i].fillna(df[i].mode()[0])
df.isna().sum()
```

```python
for i in df.select_dtypes(include=['float']).columns:
    df[i]=df[i].fillna(df[i].mean())
df.isna().sum()

    

```

```python
df.drop(["PassengerId", "Cabin", "Name"], axis=1, inplace = True)
df.head
```

```python
col = df.select_dtypes(['object']).columns
le=LabelEncoder()
for i in col:
    df[i]=le.fit_transform(df[i])
df.head()
```

```python
x = df.drop("Transported", axis=1)
y = df.Transported
```

```python
trainx, testx, trainy, testy = train_test_split(x, y, test_size = 0.3)
```

```python
dt = DecisionTreeClassifier()
model_dt = dt.fit(trainx, trainy)
pred_dt = model_dt.predict(testx)
print(confusion_matrix(testy, pred_dt))
print(accuracy_score(testy, pred_dt))
```

```python
knn = KNeighborsClassifier()
model_knn = knn.fit(trainx, trainy)
pred_knn = model_knn.predict(testx)
print(confusion_matrix(testy, pred_knn))
print(accuracy_score(testy, pred_knn))
```

```python
nb = GaussianNB()
model_nb = nb.fit(trainx, trainy)
pred_nb = model_nb.predict(testx)
print(confusion_matrix(testy, pred_nb))
print(accuracy_score(testy, pred_nb))
```

```python
lr = LogisticRegression()
model_lr = lr.fit(trainx, trainy)
pred_lr = model_lr.predict(testx)
print(confusion_matrix(testy, pred_lr))
print(accuracy_score(testy, pred_lr))
```
