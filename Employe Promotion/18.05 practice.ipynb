```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,VotingClassifier,StackingClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
```

```python
df = pd.read_csv("C:/Users/gpro0/OneDrive/Рабочий стол/Data Science course/employee_promotion.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
for i in df.select_dtypes(['object']).columns:
    df[i] = df[i].fillna(df[i].mode()[0])
df.isna().sum()
```

```python
for i in df.select_dtypes(include=['float']).columns:
    df[i]=df[i].fillna(df[i].mean())
df.isna().sum()
```

```python
df.drop("employee_id", axis=1, inplace=True)
df.head()
```

```python
col = df.select_dtypes(['object']).columns
le=LabelEncoder()
for i in col:
    df[i]=le.fit_transform(df[i])
df.head()
```

```python
x=df.drop('is_promoted',axis=1)
y=df.is_promoted
```

```python
trainx,testx,trainy,testy=train_test_split(x,y,test_size=0.2)
```

```python
rf=RandomForestClassifier(n_estimators=27, criterion="log_loss")
model=rf.fit(trainx,trainy)
pred=model.predict(testx)

print(accuracy_score(testy,pred))
print(confusion_matrix(testy,pred))
```

```python
dt=DecisionTreeClassifier(max_depth=150, criterion="log_loss")
lr=LogisticRegression()
knn=KNeighborsClassifier(n_neighbors=21, n_jobs=-1)
nb=GaussianNB()
estimators=[('DT',dt),('LR',lr),('RF',lr),('KNN',knn),('NB',nb)]
vt=VotingClassifier(estimators, flatten_transform= True)
st = StackingClassifier(estimators)
```

```python
for i in [lr, dt, nb, knn, vt, rf, st]:
    model = i.fit(trainx, trainy)
    pred = model.predict(testx)
    print(f"{str(i)}\n,{confusion_matrix(testy, pred)}\n,{accuracy_score(testy, pred)}\n-----------")
```

```python
ab = AdaBoostClassifier()
model_ab=ab.fit(trainx,trainy)
pred_ab=model_ab.predict(testx)

print(accuracy_score(testy,pred_ab))
print(confusion_matrix(testy,pred_ab))
```

```python
gb = GradientBoostingClassifier()
model_gb=gb.fit(trainx,trainy)
pred_gb=model_gb.predict(testx)

print(accuracy_score(testy,pred_gb))
print(confusion_matrix(testy,pred_gb))
```
